```python
%matplotlib widget
import pandas as pd
import numpy as np

import datetime
```

```python
rocket_df = pd.read_csv(r'C:\Users\novak\OneDrive\Projekti\Raketa\2020-11-14 Video in podatki senzorjev\rocket-data-12-5.csv', sep=';') 
# rocket_df['Timestamp'] = rocket_df['Timestamp'].apply(lambda x: datetime.datetime.strptime(x, '%Y-%m-%d %H:%M:%S.%f').time())
rocket_df['Timestamp'] = pd.to_datetime(rocket_df['Timestamp'], format='%Y-%m-%d %H:%M:%S.%f')
rocket_df['Timestamp']
```

```python
rocket_df.dtypes
```

```python
rocket_df = rocket_df.set_index('Timestamp')
rocket_df
```

```python
rocket_df['Velocity'] = 0.0
rocket_df.shape
for r in range(1, rocket_df.shape[0]):
    rocket_df['Velocity'][r] =  (rocket_df['Altitude'][r] - rocket_df['Altitude'][r-1]) / (rocket_df.index[r] - rocket_df.index[r-1]).microseconds * 1000000
rocket_df
```

```python
rocket_df_resampled = rocket_df.resample('100ms').mean()
rocket_df_resampled
```

```python
import matplotlib.dates as mdates
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(1,1,1)  
plt.plot(rocket_df_resampled['Ay'][0:400])
plt.plot(rocket_df_resampled['Altitude'][0:400])
plt.plot(rocket_df_resampled['Velocity'][0:400])
plt.xticks(rotation='vertical')
ax.xaxis.set_major_locator(mdates.MicrosecondLocator(interval=100000))
plt.grid()
plt.show()
```
